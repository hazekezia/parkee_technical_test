```python
import requests
import pandas as pd
```

```python
url = f'http://universities.hipolabs.com/search?country=United%20States'
response = requests.get(url)
universities_data = response.json()
```

```python
# Convert to DataFrame
df = pd.DataFrame(universities_data)
df
```

```python
# Filter out rows where 'state-province' is missing
df = df[df['state-province'].notna()]
df
```

```python
# Select relevant columns
df = df[['name', 'web_pages', 'country', 'domains', 'state-province']]
df
```
